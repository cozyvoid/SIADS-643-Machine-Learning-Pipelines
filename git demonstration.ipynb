> `git status` (what is going on)

> now go to Console > File > New File > example name: "main.py" (Tab: `main.py U`)
> git does not track files unlessyou explicitly tell it to, so `git add main.py`
> If you use `git status`, main.py now comes up with a "Changes to be committed:" with "main.py" listed and Tab: `main.py A`

> `git commit main.py - m 'Add main!'` <-- The A in Tab: "main.py A" goes away, also no longer green.

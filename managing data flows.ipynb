# Managing Data Flows 

---

### What is a pipeline?
## (Software level pipelines)

- Set of steps 
- Repeatable
- Isolated from each other
- With a configuration around how the steps are executed.
- A pipeline is a DAG (Directed Acyclic Graph)


### What is a DAG?

- A term Directed Acyclic Graph refers to a network of items whose relationshuips are directional, and where those relationships don't form any loops.



### DAG's

- Data processing flows are often referred to as DAG's
- They have a series of directed steps
- Each step can have one or more dependencies
- Each step can produce one or more outputs
- Steps should not be arranged to form circular dependencies

### Paradigms

- Imperative:
    - A programming paradigm that uses control structures to describe **how** an objective is achieved
        - Ex:
            - Python
            - C
            - Jav, etc.
- Declarative:
    - A programming paradigm that defines an end condition to describe **what** the objective is.
        - Ex:
            - Makefile
            - DVC
            - Ansible, etc.


### DAG's and declarative tools

- DAG's work well with declarative tools
- You can leave the work of checking dependencies and deciding when to recreate them to the tool by sufficiently describing your end state.

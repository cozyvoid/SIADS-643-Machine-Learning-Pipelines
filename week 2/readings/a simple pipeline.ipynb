# A Simple Pipeline

---

### Pipelines 

- A pipeline does not have to be extremely complicated.
- But there are reasons to not be too simple. 

### `pipeline.py`



- Block 1: Dependency imports 
- Block 2 + 3: Read and clean data 
- Block 4: Split the data 
- Block 5: Train a model 
- Block 6: Display metrics 
- Block 7: Save the trained model 

#### A note about `pickle`



- `pickle` is a handy tool to serialize (save) and deserialize (read) Python objects like functions and classes.
- But since it allows you to load a full-fledge python object from a file, you must trust the source of that file. 

### Weaknesses 



> Inflexible, tightly coupled, slow parts 

### Tooling 

- Simple pipelines can be helpful for exploratory analysis, but they are hard to scale. 
- As your data and analyses grow having to run the whole pipelines with each change can be a very slow process.
- Consider using tooling that helps with these challenges, rather than building our own.

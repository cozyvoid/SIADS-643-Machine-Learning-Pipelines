### You are building a machine learning pipeline for student admission data. Your goal is to try and predict if a given student will come to the University.

### The following are the fields you are given from the University system. This is data from the last year you are going to use to train your model. In production, your model will run right after acceptance to the University is granted. We are trying to predict "student_matriculated"

### Date of creation: The date the student was added to the system. (Normally this is the date they submit their application on, however, if this is a graduate school application, and the student submitted an undergrad app, this would be the date the undergrad application was submitted)

```
studentid: The student's unique identifier

date_of_acceptance: The date the student was accepted.

Firstname: The first name of the student

Lastname: The last name of the student

total_paid: For the lifetime of the student account, what is the total amount paid.

Home_address: The home address (address1, address2, city, state, zip) of the student. This address has been validated by the post office.

student_phone: The 10 student's contact phone number.

student_application_gpa: The GPA of the student as listed in the application

student_matriculated: Did the student accept the offer to come. True or False
```

### What values should you drop to make sure you don't have data leakage in your pipeline?

- [ ] `data_of_acceptance`
- [x] `total_paid`
- [ ] `student_application_gpa`
- [ ] `first_name`
- [ ] `last_name`

### Starting off

- open up your terminal in vscode or locally

### Running `pwd` will tell you what directory you are in

- `pwd`

### `git init`

- `git init`
- this will create a repository for you. You only need to do this once per project.


### `git remote`

- when you want to share your code, you need to add a remote
- you only need to do this once per project
- after creating you repo online, you should see a line that starts off with

- `git remote add origin XXXXX`

- paste it in
- (ex: `git remote add origin` github.com/mlhess/week1)

### tell git who you are

- you need to run these once. Make sure to replace with your information.
- `git config --global` user.email
"you@example.com"
- ` git config --global user.name "Your name"`

### Do some work.

- create some files in this directory. Put some content in the files. 
- (on the command line you can use the touch command to create a file)
- (touch filename)

### `git status `

- `git status` will provide the status of the current repo

### `git add`   

- `git add` will tell git to add a file to the repo
- `git add <filename>`

### `git commit`

- `git commit <filename> -m <message>`
- Commit the filename to the repo with the change message specified.

### `git push`

- `git push origin`
- takes all changes you have committed and sends them to the remote server
- after you run this, hyou can look at github and see the changes

### `flow`

- normally you do work >>
- you commit your work >>
- you do more work >>
- you commit your work ->
- when done you push 

### `git log`

- the command git log will show you all the changes you have made

![git hash](<attachment:Screenshot 2026-07-01 165412.png>)

> notice the long string after the word commit?
> this is the commit *hash*

### `git revert`

- git revert <hash> --no-edit
- will undo the commit as designated by the hash
- you can get the hash from git log

### `git branch`

- a branch starts off as a copy of the code you branch from.
- you can make changes in a branch without affecting other branches.

### syntax

- `git branch <new branch>` * Creates a branch
- `git branch` * provides info on branches
- `git checkout branch`

# Data Leakage 

---

### Data Leakage 

- Data leakage is when data gets into your model that your model would not have access to

### Common Causes

- Feature engineering
    - Your dataset has feature that it should not
- Improper cross validation 
    - Your gets evaluated on data it has already been trained on
- Data processing steps
    - Early normalization of data 

## Feature Importance

### Eval Scores 

- How are you  evaluating your model, does this number change over the life of your pipeline?
- Is the number too good (or too bad) to be true?

### 1. Scenario: You're training a machine learning model to predict future sales. You include next month's sales figures as one of the input features. Is this data leakage?

- Yes

### 2. Scenario: In a model predicting credit card fraud, you use the customer's past transaction history. Is this an example of data leakage?

- No

### 3. Scenario: In a model to predict movie ratings, you use user reviews that were written after watching the movie. Is this data leakage?

- Yes

### 4. Scenario: For a weather prediction model, you use historical weather data along with future climate predictions from another model. Is this an example of data leakage?

- Depends on the other model's features 

### 5. Scenario: You are developing a predictive model for the success of startups in securing venture capital funding. The model factors in the startup's growth rate, market trends, and founder experience. Additionally, it uses data from a network analysis tool that maps relationships between startups and investors. The network data is historical and static, representing the state before any current fundraising rounds begin. Therefore, none of the relationships are formed or strengthened during the fundraising rounds within the prediction timeframe. Is this an example of data leakage?

- No, since the network data is historical and does not change during the prediction period.

### DAG's with the make command

- The `make` command is traditionally used in software compilation pipelines.
- Usually available on Unix/Linux systems with software build packages installed.
- Its flow is defined by **Makefile** which by convention is stored with an application's source code.
- Very commonly used with C/C++ projects to manage compilation and linking work flows
- The Makefile defines:
    - "Targets"
    - The dependenciies of each target
    - If necesssary, the command needed to create a target
- It uses the file system time stamps on the dependencies and targets to only recreate targets which are older than thier dependencies
- The declarative style of the Makefile lets us focus on the logic of each step
- We can leave the burden of managing the DAG to the `make` command

### What problems do we solve?

### Problems 

- code revisions
- code collaborations 
- change history 
- last known good 
- refactoring a code base while still working on the current code base 
- accountability (who changed what and why)
- auditing 

#### code revisions 

- the core of any version control system is to track changes to code. 
- what changed, in what file, by who and when was it changed.

#### working with co-workers 

- How can you talk about and propose changes. 
- Github was the first website that made It popular, but the underlying technology is git. 
- Git has a "merge" feature that allows you to merge others changes (or your own) into a working branch.
- Git breaks code into chunks, so 2 (or more) people can edit the same file and most of the time it will work (When it doesn't, that is called a merge conflict)

#### Last known good 

- It is helpful to know what version of code is working. Git allows you to tag code. The tag can be shared.

#### refactor code 

- Git has a feature known as branches that is helpful for refactoring code or working on code before it is ready for release.

#### accountability and auditing 

- If you are using Git correctly, you can check a large number of your auditing and accountability settings.
